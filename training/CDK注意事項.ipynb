### スタック削除時はリソースが削除されているか確認  
CDKでリソースを作成した場合、デフォルトではスタック削除時にリソースが削除されない。この状態でデプロイとデストロイを繰り返すと不要なリソースが増え続けるので注意が必要

※「pythonstack」という名前のスタックで作成したS3が、スタック削除後も残っている
![スタックが削除されてもリソースが残る](./img/スタック削除後にリソースが残る.png)

### スタックの識別名とリソース名は異なる  

CDKでリソース作成時に指定する名称はCloudFormation上の論理名として設定される。  

リソース名が異なっていれば、リソース名が同じでも正常なコードを判定されるため修正もれには要注意
![リソース名が重複](./img/リソース名が重複.png)
